```python
top_countries = soup.find_all('li', class_='top-country')

# Print the number of elements found to debug
print(f"Found {len(top_countries)} elements with class 'top-country'")

# Iterate through each <li> element and extract the <a> tag details
for country in top_countries:
    a_tag = country.find('a')  # Find the <a> tag within the <li> element
    if a_tag:  # Check if the <a> tag exists
        country_name = a_tag.get_text()  # Get the country name (link text)
        country_url = a_tag['href']  # Get the URL (href attribute of <a> tag)
        print(f"{country_name}: {country_url}")
```
