```python
import pandas as pd
```

```python
df_marketcap = pd.read_csv("Companies ranked by Market Cap.csv")
```

```python
df_comapnies = pd.read_csv("Companies_list.csv")
```

```python
df_comapnies.head()
```

```python
import pandas as pd


# Split the 'Company' column at '(' and expand the result
split_columns = df_comapnies['Name'].str.split('(', expand=True)

# Assign new columns to the original dataframe
df_comapnies['Company Name'] = split_columns[0]
df_comapnies['Ticker'] = split_columns[1].str.replace(')', '', regex=False)

```

```python
df_comapnies.head()
```

```python
df_marketcap.head()
```

```python
merged_df = pd.merge(df_comapnies, df_marketcap, left_on='Ticker', right_on='Symbol', how='left')
```

```python
merged_df.head()
```

```python
merged_df = merged_df[['Name_x', 'Exchange', 'SuperSector', 'Country', 'Rank', 'marketcap']]
merged_df = merged_df.rename(columns={"Name_x": "Name"})
merged_df = merged_df.sort_values(by='marketcap', ascending=False)
merged_df.reset_index(drop = True, inplace = True)
```

```python
lang_df = pd.read_csv("Lang_lookup.csv")
```

```python
#Merge the langauge 
merged_df_new = pd.merge(merged_df, lang_df, left_on='Country', right_on='Country', how='left')
```

```python
merged_df_new['Language Code'] = merged_df_new['Language Code'].fillna('lang_en')
```

```python
merged_df_new.drop_duplicates(inplace = True)
merged_df_new.reset_index(drop = True, inplace = True)
```

```python
merged_df_new = merged_df_new.drop_duplicates(subset=['Name', 'Language Code'])
```

```python
merged_df_new.reset_index(drop = True, inplace = True)
```

```python
merged_df_new.head(200)
```

```python
merged_df_new.to_csv('Companies_list_enhanced.csv', index=False)
```
